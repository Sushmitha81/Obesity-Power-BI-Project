# Pre Processing

### Drop Duplicates

```python
Obesity_Data.duplicated(keep='first').sum()
```

```python
Obesity_Data=Obesity_Data.drop_duplicates()
```

### Drop Unwanted Columns

```python
Obesity_Data.drop("DOB",axis=1,inplace=True)
```

```python
Obesity_Data.head()
```

 ### Drop Null Values

```python
Obesity_Data.isnull().sum()
```

```python
(Obesity_Data.isnull().sum()/Obesity_Data.shape[0]*100).sort_values(ascending=False)
```

#### 'Colour' has 88% null records so lets drop the column

```python
Obesity_Data.drop('Colour',axis=1,inplace=True)
```

```python
(Obesity_Data.isnull().sum()/Obesity_Data.shape[0]*100).sort_values(ascending=False)
```

#### we have 8% to 25% null values so we are filling the numerical columns with mean and categorical columns with mode

```python
for i in Obesity_Data.columns:
    if Obesity_Data[i].dtype =='float64':
        Obesity_Data[i].fillna(Obesity_Data[i].mean(),inplace=True)
    else:
        Obesity_Data[i].fillna(Obesity_Data[i].mode()[0],inplace=True)
```

```python
(Obesity_Data.isnull().sum()/Obesity_Data.shape[0]*100).sort_values(ascending=False)
```

#### Rounding of values and converting the height from meters to centimeters

```python
for i in Obesity_Data.columns:
    if Obesity_Data[i].dtype == "float64":
        Obesity_Data = Obesity_Data.round(1)
Obesity_Data.head()
```

```python
Obesity_Data["Height"]=Obesity_Data["Height"]*100
```

```python
Obesity_Data.head()
```

#### Converting Float to int

```python
for i in Obesity_Data.columns:
    if Obesity_Data[i].dtype == "float64":
        Obesity_Data[i] = Obesity_Data[i].astype('int64')
Obesity_Data.head()
```
